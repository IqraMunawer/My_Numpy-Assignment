# Q2

```python
path = r'Assignment/States/'                     # use your path
all_files = glob.glob(os.path.join(path, "*.csv"))   
# advisable to use os.path.join as this makes concatenation OS independent
df_from_each_file = (pd.read_csv(f) for f in all_files)
us_census  = pd.concat(df_from_each_file, ignore_index=True)
us_census
```

# Q6

```python
pop = us_census.GenderPop.str.split('_', expand = True)
us_census[['Male', 'Female']] = pop
us_census
```

# Q7

```python
us_census['Male'].replace('(M)','',regex = True, inplace = True)
us_census['Female'].replace('(F)','',regex = True, inplace = True)
us_census[['Male', 'Female']] = us_census[['Male', 'Female']].apply(pd.to_numeric, errors='coerce')
us_census[['Male', 'Female']].dtypes
```

```python
Fdata = us_census[['Female', 'Income']]
Fdata
plt.scatter('Income', 'Female', data = Fdata) 
plt.show()
```
